```python
import pandas as pd
import numpy as np

a = 'https://www.czso.cz/documents/10180/32853391/300002164403.xlsx/1fed9fb2-4dbf-4b0f-bfbe-4c2d3202a18f?version=1.1'
     
data = pd.read_excel(a, index_col = 0, skiprows = 5)
data = data.drop(data.columns[[1,3,5,7,9,11,13,15]], axis=1)

col1 = data.iloc[0].fillna(method = 'ffill')
col2 = data.iloc[2].str.split(pat='/ years', n = 0).str.get(0).str.split(pat='/ 65 years or more', n = 0).str.get(0).fillna(method = 'ffill')
col3 = data.iloc[3].fillna(method = 'ffill')



columns = [col1, col2, col3]
columns = list(zip(*columns))
data.columns = columns

data = data.drop(data.columns[[-1]], axis=1)


letadlo = pd.read_excel(a, index_col = 0, skiprows = 5)
letadlo = letadlo.drop(letadlo.columns[[0,2,4,6,8,10,12,14,16]], axis=1)
letadlo = letadlo.drop(letadlo.index[[0,1,2,3,4,7,-1,-2,-8]])

letadlo.columns = pd.MultiIndex.from_tuples(data.columns, names=['Země','',''])

letadlo = letadlo.rename(index={'Lucembursko1)':'Lucembursko'})

letadlo
```
